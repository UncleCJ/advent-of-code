```python
part1_example_input = '''###############
#...#...#.....#
#.#.#.#.#.###.#
#S#...#.#.#...#
#######.#.#.###
#######.#.#...#
#######.#.###.#
###..E#...#...#
###.#######.###
#...###...#...#
#.#####.#.###.#
#.#...#.#.#...#
#.#.#.#.#.#.###
#...#...#...###
###############'''
```

```python
def parse_input(lines):
    results = {
        'start': (-1, -1),
        'end': (-1, -1),
        'track': set(),
        'walls': set(),
        'size': len(lines)
    }
    for row, line in enumerate(lines):
        assert(len(line) == results['size'])
        if row in (0, results['size']-1):
            continue
        for col, tile in enumerate(line):
            if col in (0, results['size']-1):
                continue
            coord = (row, col)
            match tile:
                case '.' | 'S' | 'E':
                    results['track'].add(coord)
                    if tile != '.':
                        if tile == 'S':
                            results['start'] = coord
                        else:
                            results['end'] = coord
                case '#':
                    results['walls'].add(coord)
                case _:
                    raise ValueError
    return results

parsed_input = parse_input(part1_example_input.splitlines())
```

```python
def find_options(design_tail, available):
    return [(a, design_tail[len(a):]) for a in available if design_tail.startswith(a)]

find_options(parsed_input['designs'][0], parsed_input['available'])
```
