```python
with open('input.txt', 'r') as inp:
    inputdata = [line.strip() for line in inp.readlines()]
```

```python
from itertools import permutations  

answer = 0

testdata = """35
20
15
25
47
40
62
55
65
95
102
117
150
182
127
219
299
277
309
576""".splitlines()

l_preamble = 25
data = [int(x) for x in inputdata]

window = data[0:l_preamble]

for i_inv, invalid in enumerate(data[l_preamble:], start=l_preamble):
    if invalid not in list(map(lambda t: t[0] + t[1], 
            filter(lambda t: t[0] <= t[1], permutations(window,2))
        )):
        break
    window = window[1:l_preamble] + [invalid]

print(f'{invalid} (element {i_inv}) cannot be summed from two of {window}')

for i_f, f in enumerate(data[:i_inv]):
    i_l = i_f+1
    cumulative = f
    while (i_f < i_inv and cumulative < invalid):
        cumulative += data[i_l]
        i_l += 1
        if cumulative == invalid:
            break
    if cumulative == invalid:
        break

print(f'first: {f} (index {i_f}), last: {data[i_l]} (index {i_l}), elements: {data[i_f:i_l]}, sum: {sum(data[i_f:i_l])}')
print(f'{min(data[i_f:i_l])} + {max(data[i_f:i_l])} = {min(data[i_f:i_l])+max(data[i_f:i_l])}')
```
